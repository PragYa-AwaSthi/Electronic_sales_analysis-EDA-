A.3.Determine the best time of the day to display an advertisement in a TV channel so that the probability of customers buying a product is maximum. 

```python
#spliting the timestamp and creating a new column only for time
df_annual['Time'] = df_annual['Order Date'].apply(lambda x: x.split(' ')[1])
```

```python
#further spliting hrs and minutes
df_annual['Mins'] = df_annual['Time'].apply(lambda x: x.split(':')[0])
df_annual[['Mins']].info()
```

```python
df_annual['Mins'] = df_annual['Mins'].apply(lambda x: int(x)) #converting object type to integer type
df_annual[['Mins']].info()
```

```python
df_annual['Mins'] = df_annual['Mins']*60 #converting into minutes
```

```python
#spliting and saving the minute side time in new variable
df_annual['Mins_1'] = df_annual['Time'].apply(lambda x: x.split(':')[1]) 
df_annual['Mins_1'] = df_annual['Mins_1'].apply(lambda x: int(x)) #converting into integer type
```

```python
#adding total minutes
df_annual['Time in mins'] = df_annual['Mins'] + df_annual['Mins_1'] 
```

```python
df_annual.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df_annual['Time in mins'], bins = 10, hist_kws = dict(edgecolor = 'black'))
plt.show()
```
